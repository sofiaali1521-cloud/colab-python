```python
lst = [1 , 2 , 3 , 'hello' , 'c' , [1 , 2]]
print(lst)

#append() take single value only(sabhi value ko as a single value treat krga
# = list ko aak value ki tarh treat krega)
lst.append([2 , 3 , 4])
print(lst)
#extend() ye seperate elememts add krta hai
lst.extend([2 , 3 , 4])
print(lst)
#insert(index, value)
lst.insert(2 , "value")
#pop()
lst.pop()
print(lst)
#remove() first occurance remove krta hai only
lst.remove(2)
print(lst)
#index()
lst.index(2)
print(lst.index(3))
#count()
print(lst.count(3))
```

```python
#tuples (do not allow changes ->. can not add or remove the value, we can not sort but
# can be sorted)
tp = (1,24,3,4)
a = sorted(tp)
print(a)
#[1, 3, 4, 24] ye ans hoga joki list h agar ans bhi tuple me chiye to pir ->
a = tuple(sorted(tp))
print(a)
#(1, 3, 4, 24) ye hoga ans jo ki aak tuple h

#tuple fn :-

#catergory are ->
#count values, find, size, total, highest/lowest, sort

#fns are ->
#count(), index(), len(), sum(), max()/min(), sorted()(return list)

tup = (2, 45, 35, 87, 98, 2)
print(tup.count(2))
print(tup.index(35))
print(len(tup))
print(sum(tup))
print(max(tup))
print(min(tup))
b=sorted(tup)
print(b)
```
